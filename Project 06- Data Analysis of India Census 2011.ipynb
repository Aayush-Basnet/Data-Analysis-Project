## Gender Analysis

```python
df.head()
```

```python
# total number of males by state
total_male_population_by_state = df.groupby('State_name')['Male'].sum()
total_male_population_by_state
```

```python
# total number of females by state
total_Female_population_by_state = df.groupby('State_name')['Female'].sum()
total_Female_population_by_state
```

```python
# gender ratio by state
# gender ratio(number of females per 1000 males)
gender_ratio_by_state = (total_Female_population_by_state/total_male_population_by_state)*1000
gender_ratio_by_state.sort_values(ascending =False)

# Kerala and Pondicherry has more number of female compared to male.
```

```python
Literacy_rates=(df.groupby('District_name')['Literate'].sum()/df.groupby('District_name')['Population'].sum())*100
Literacy_rates.sort_values(ascending=False)

# Pathanamthitta district has highest literacy rate with 88.75% followed by Kottayam with 88.62%.
```
