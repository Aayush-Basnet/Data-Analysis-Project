## Price Analysis

```python
df.head()
```

```python
df.insert(1,'year',df.date.dt.year)
df.head()
```

```python
group_london=df[df['area']=='city of london']
```

```python
df[(df['year']==1995) & (df['area']=='city of london')]
```

```python
group_london.groupby('year').average_price.describe()
```

```python
group_london.groupby('year').average_price.max()
```

```python
group_london.groupby('year').average_price.max().plot(kind = 'line',title='Price History',xlabel='Year',ylabel='Max Price Per Year')
```

```python
group_london.groupby('year').average_price.mean()
```

```python
group_london.groupby('year').average_price.mean().plot(kind = 'line',title='Price History',xlabel='Year',ylabel='Mean Price Per Year')
```
